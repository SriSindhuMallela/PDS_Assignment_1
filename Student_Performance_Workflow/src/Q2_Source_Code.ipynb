# data cleaning part:

```python
import pandas as pd

# Specifying the path to my raw data CSV file
file_path = "C:/Users/srisi_w/Desktop/Python_Practice/Student_Performance_Workflow/raw data/student_performance.csv"  

# Loading the CSV file into a Pandas DataFrame
df = pd.read_csv(file_path)


# Checking for missing values
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)
```

### saving cleaned data:

```python
# Saving the cleaned data of above
df.to_csv("C:/Users/srisi_w/Desktop/Python_Practice/Student_Performance_Workflow/clean_data/cleaned_data.csv", index=False)
```

```python
# Visualization 1: Histogram of Math Scores
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x="math score", bins=20)
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.title("Distribution of Math Scores")
plt.savefig("C:/Users/srisi_w/Desktop/Python_Practice/Student_Performance_Workflow/results/Visualisations/Visualisationsmath_scores_histogram.png")
plt.show()
```

```python
# Visualization 2: Bar Chart of Gender Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution")
plt.savefig("C:/Users/srisi_w/Desktop/Python_Practice/Student_Performance_Workflow/results/Visualisations/gender_distribution_bar_chart.png")
plt.show()
```

```python
# Visualization 3: Scatter Plot of  Writing Score vs. Math Score
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="writing score", y="math score")
plt.xlabel("Writing Score")
plt.ylabel("Math Score")
plt.title(" Writing Score vs. Math Score")
plt.savefig("C:/Users/srisi_w/Desktop/Python_Practice/Student_Performance_Workflow/results/Visualisations/writingscore_vs_math_score_scatter.png")
plt.show()
```

```python
# Visualization 4: Box Plot of Parental Education vs. Math Score
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="parental level of education", y="math score")
plt.xticks(rotation=45, ha="right")
plt.xlabel("Parental Education Level")
plt.ylabel("Math Score")
plt.title("Parental Education vs. Math Score")
plt.tight_layout()
plt.savefig("C:/Users/srisi_w/Desktop/Python_Practice/Student_Performance_Workflow/results/Visualisations/parental_education_vs_math_score_boxplot.png")
plt.show()
```

```python
# Visualization 5: Countplot of Test Preparation Course
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="test preparation course")
plt.xlabel("Test Preparation Course")
plt.ylabel("Count")
plt.title("Test Preparation Course")
plt.savefig("C:/Users/srisi_w/Desktop/Python_Practice/Student_Performance_Workflow/results/Visualisations/test_preparation_course_countplot.png")
plt.show()
```

```python
#visualization 6: Pie Chart of number of students who completed and did not complete the test preparation course

import pandas as pd
import matplotlib.pyplot as plt


# Count the number of students who completed and did not complete the test preparation course
test_preparation_counts = df['test preparation course'].value_counts()


# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(test_preparation_counts, labels=test_preparation_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Test Preparation Course Completion")
plt.savefig("C:/Users/srisi_w/Desktop/Python_Practice/Student_Performance_Workflow/results/Visualisations/test_preparation_course_pie_chart.png")
plt.show()
```
